```python
import os
import torch
import json
import soundfile as sf
from qwen_tts import Qwen3TTSModel
from qwen_tts.inference.qwen3_tts_model import VoiceClonePromptItem
from transformers import BitsAndBytesConfig
#from faster_qwen3_tts import FasterQwen3TTS as Qwen3TTSModel                                                                                                                           
import inspect

torch.cuda.empty_cache()
torch.backends.cuda.sdp_kernel(enable_flash=False, enable_math=True, enable_mem_efficient=True)


bnb_config = BitsAndBytesConfig(
  load_in_8bit=True,
  llm_int8_threshold=6.0,
  llm_int8_has_fp16_weight=False,
  llm_int8_skip_modules=["lm_head", "embed_tokens"]
)

model = Qwen3TTSModel.from_pretrained('Qwen/Qwen3-TTS-12Hz-0.6B-Base', device_map ='cuda:0', quantization_config=bnb_config)


voice_clone_prompt = model.create_voice_clone_prompt(
  ref_audio='/workspace/data/arteaga/output2.wav',
  ref_text='/workspace/data/arteaga/output2.json',
  x_vector_only_mode=False
)


print(voice_clone_prompt)
```

```python
ref_code = voice_clone_prompt[0].ref_code
```

```python
from torch.utils.tensorboard import SummaryWriter

writer = SummaryWriter(log_dir="/runs/voice_clone") 
writer.add_embedding( ref_code, metadata=[f"step_{i}" for i in range(ref_code.size(0))], tag="ref_code_steps" ) 
writer.close()
```
